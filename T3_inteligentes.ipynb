```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
import scipy.stats as stats
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
import pandas as pd
```

```python
from google.colab import files

data = pd.read_csv('dataset.csv') 
```

```python
data
```

```python
X = data[["platelets","creatinine_phosphokinase"]].to_numpy()
x_zscore = stats.zscore(X, axis=0)
X = np.hstack((np.ones([100,1]),x_zscore))
y = data[["age"]].to_numpy()
```

```python
X_T = np.transpose(X)
inversa = np.linalg.inv(np.dot(X_T, X))
mult = np.dot(X_T, y)
w = np.dot(inversa,mult)
```

```python
w
```

```python
 y_pred = np.dot(X,w)
```

```python
from sklearn.metrics import mean_squared_error

for i in range(100):
  mse.append(mean_squared_error(y[i], y_pred[i]))
  mae.append(np.mean(np.abs(y[i]-y_pred[i])))
```

```python
fig, axs= plt.subplots(nrows=1, ncols=2, figsize=(24, 7))
axs[0].plot(y_pred, color = 'blue')
axs[0].plot(y ,color = 'red')
axs[0].legend(["predito","real"],loc="best")

axs[1].plot(mse, color = 'green')
axs[1].plot(mae, color = 'red')



     
```
